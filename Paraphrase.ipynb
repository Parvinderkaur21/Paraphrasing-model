## ***PARAPHRASING TEXTS USING PYTHON*** 

The task is to build a model which can paraphrase any sentence. 

We can paraphrase text in python by using Parrot library. Parrot is a paraphrase based utterance augmentation framework purpose built to accelerate training NLU models. A paraphrase framework is more than just a paraphrasing model.

The paraphrasing technology of the Parrot library is created on the T5 algorithm. T5 stands for "text to text transformer".

```python
from parrot import Parrot
import torch
import warnings
warnings.filterwarnings("ignore")
```

*Now , setting reproducibility of the paraphrasing.*

```python
def random(seed):
    torch.manual_seed(seed)
    
    if torch.cuda.is_available():
        torch.cuda.manual_seed_all(seed)
```

```python
random(1234)
```

*Load the text paraphrasing model*

```python
parrot = Parrot(model_tag="prithivida/parrot_paraphraser_on_T5", use_gpu=False)
```

*Providing the input text*

```python
text = ["Which course should I take to get started in data science?"]
```

*Generating the paraphrased text*

```python
for phrase in text:
    print("-"*100)
    print("Input_phrase: ", phrase)
    print("-"*100)
    para_phrases = parrot.augment(input_phrase=phrase)
    for para_phrase in para_phrases:
        print(para_phrase)
```
